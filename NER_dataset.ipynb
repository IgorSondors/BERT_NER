```python
def data_1C(pth1, pth2, pth3, pth4):
    df1 = pd.read_excel(pth1)
    df1 = df1.drop(index=0)

    df1['тип'] = ['Колобокс_Шины' for i in range(len(df1))]
    df1 = df1.rename(columns={'Ширина': 'Шир', 'Диаметр': 'Диам', 'Профиль': 'Проф', 'Марка (бренд)': 'Марка'})

    df2 = pd.read_excel(pth2)
    df2['тип'] = ['легковая_резина' for i in range(len(df2))]

    df3 = pd.read_excel(pth3)
    df3['тип'] = ['грузовая_резина' for i in range(len(df3))]

    df4 = pd.read_excel(pth4)
    df4['тип'] = ['диски' for i in range(len(df4))]

    return df1, df2, df3, df4

def concat_and_clear(lst):
    df = pd.concat(lst, ignore_index=True)
    rm = ['Сезонность', 
        'Артикул', 
        'Код 1С', 
        'Код_произв',  
        'РРЦ', 
        'Тип_диска', 
        'Отв', 
        'PCD', 
        'DPCD', 
        'Вылет', 
        'Ступица',
        'RN', 
        'Остаток', 
        'Резерв', 
        'Заказ', 
        'Цена_руб',
        'Распр', 
        'Происх', 
        'Сезон',
        'Скор',
        'Размер',
        'Шип']

    df = df.drop(columns=rm)
    df = df.rename(columns={'Шир': 'Ширина', 'Диам': 'Радиус', 'Проф': 'Высота', 'Модель': 'Линейка'})

    df = df[df['Ширина']!="@ширина рез."]
    df = df[df['Высота']!="@профиль"]
    df = df[df['Радиус']!="@диаметр"]
    df = df[df['Линейка']!="@модель дис."]
    df = df.reset_index(drop=True)

    return df

pth1 = "/home/sondors/Documents/price/BERT_NER/csv/Прайс_Колобокс_Шины_2018-07-07 (XLS).xls"
pth2 = "/home/sondors/Documents/price/BERT_NER/csv/легковая_резина.xls"
pth3 = "/home/sondors/Documents/price/BERT_NER/csv/грузовая_резина.xls"
pth4 = "/home/sondors/Documents/price/BERT_NER/csv/диски.xls"

df1, df2, df3, df4 = data_1C(pth1, pth2, pth3, pth4)
df_opensource = concat_and_clear([df1, df2, df3, df4])
df_opensource
```

```python
# Функция для проверки, содержит ли значение в "Наименование" значение из указанной колонки и округления этой колонки (если не NaN). Округление бесполезных float до целых
def process_column(row, column_name):
    if not pd.isna(row[column_name]):
        contains_column_value = str(row['Наименование']).find(str(row[column_name])) != -1
        if not contains_column_value:
            row[column_name] = round(row[column_name])
    return row

columns_to_process = ['Ширина', 'Высота', 'Радиус']

for column in columns_to_process:
    df_opensource = df_opensource.apply(process_column, args=(column,), axis=1)
df_opensource
```

```python
def process_text(input_text):
    # Добавляем пробелы к букве "R", если она идет слитно с цифрой
    pattern1 = r'(R)(\d)'
    processed_text = re.sub(pattern1, r'R \2', input_text)

    pattern2 = r'(\d)(R)'
    processed_text = re.sub(pattern2, r'\1 R', processed_text)

    # Заменяем символы "/", "\", "|" на пробелы
    # processed_text = re.sub(r'[\/\\|]', ' ', processed_text)

    # Заменяем символы "/" на " / "
    processed_text = processed_text.replace("/", " / ")
    # Заменяем символы "\" на " \ "
    processed_text = processed_text.replace("\\", " \\ ")
    # Заменяем символы "|" на " | "
    processed_text = processed_text.replace("|", " | ")

    # Убираем повторяющиеся пробелы
    processed_text = re.sub(r'  +', ' ', processed_text)
    return processed_text

df_opensource['Наименование'] = df_opensource['Наименование'].apply(process_text)
df_opensource
```

```python
df_opensource = df_opensource.rename(columns={'Ширина': 'width', 'Радиус': 'radius', 'Высота': 'height', 'Марка': 'brand', 'Наименование': 'offer', 'Линейка': 'line', 'тип': 'type'})
df_opensource
```

```python

# data = {"Наименование":"Amtel 175/65R14 82Q Nordmaster (В-228)", "Ширина":175, "Высота":65, "Радиус":14, "Марка":"Amtel", "Линейка":"Nordmaster"}
# df = pd.DataFrame(data, index=[0])

# Функция для создания BIO-тегов на основе полных совпадений
def create_bio_tags(row):
    name = row['offer']
    width = str(row['width'])
    height = str(row['height'])
    radius = str(row['radius'])
    brand = row['brand']
    line = row['line']
    
    # Создаем пустой список для хранения тегов
    tags = ['O'] * len(name.split())

    # Проверяем каждое слово в наименовании и устанавливаем соответствующий тег
    for i, word in enumerate(name.split()):
        if word == width:
            tags[i] = 'B-width'
        elif word == height:
            tags[i] = 'B-height'
        elif word == radius:
            tags[i] = 'B-radius'
        elif word == brand:
            tags[i] = 'B-brand'
        elif word == line:
            tags[i] = 'B-line'

    # Собираем теги в строку
    return ','.join(tags)

# Применяем функцию к каждой строке DataFrame и создаем новую колонку с тегами
df_opensource['BIO_Tags'] = df_opensource.apply(create_bio_tags, axis=1)
df_opensource
```
