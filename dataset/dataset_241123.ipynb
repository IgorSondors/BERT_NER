```python
import pandas as pd
import os

# Путь к папке с CSV-файлами
folder_path = '/home/sondors/Documents/price/BERT_NER/csv/Kazakov_241123/разметка'

# Список для хранения DataFrame каждого файла
dfs = []

# Чтение всех CSV-файлов в папке
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        # Чтение CSV-файла и добавление его DataFrame в список
        df = pd.read_csv(file_path, sep=';')
        dfs.append(df)

# Объединение всех DataFrame в один
merged_df = pd.concat(dfs, ignore_index=True)

# Вывод объединенного DataFrame
merged_df
```

Train/test split, concat

```python
train_dataset = merged_df.sample(frac=0.9,random_state=200)
test_dataset = merged_df.drop(train_dataset.index).reset_index(drop=True)
train_dataset = train_dataset.reset_index(drop=True)
```
